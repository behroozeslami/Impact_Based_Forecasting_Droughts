```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date 
```

```python
events = pd.read_csv('./desinventar_uganda_droughts_admin1.csv',index_col=0)
events.date = pd.to_datetime(events.date,infer_datetime_format=True)
events.head()
```

### Drought events during years

```python
events.groupby('date').sum().plot()
plt.ylabel('count');
```

### Correlation with SPEI

```python
path = '../Drought_indicators/Uganda/'
spei = pd.read_csv(path+'UG_SPEI.csv',parse_dates=True)
spei.Date = pd.to_datetime(spei.Date,infer_datetime_format=True)
```

```python
events_spei = events.join(spei.set_index(['Date','districts'])['SPEI'],on=['date','district'])
```

```python
events_spei.head()
```

```python
sns.boxplot(x='drought',y='SPEI',data=events_spei);
```

```python
bins = np.arange(-5,5,0.5)
sns.distplot(events_spei[events_spei.drought==False].SPEI.dropna(),bins=bins,norm_hist=True,label='drought = False')
ax=plt.gca()
sns.distplot(events_spei[events_spei.drought==True].SPEI.dropna(),bins=bins,norm_hist=True,ax=ax,label='drought = True')
plt.legend();
```

```python
events_spei.dropna().groupby('date').agg({'drought':sum,'SPEI':np.nanmean}).plot()
```

### Correlation with NDVI 

```python
NDVI = pd.read_csv(path+'UG_NDVI_adm_1.csv',sep=';',usecols=['ADM1_EN','Month','Year','ndvi_mean'])
NDVI.rename(columns={'ADM1_EN':'district','ndvi_mean':'ndvi'},inplace=True)
NDVI['date'] = NDVI.apply(lambda x:date(x.Year,x.Month,1),axis=1)
NDVI.drop(labels=['Month','Year'],axis=1,inplace=True)
#NDVI.Date = pd.to_datetime(spei.Date,infer_datetime_format=True)
```

```python
events_ndvi = events.join(NDVI.set_index(['date','district']),on=['date','district'])
```

```python
sns.boxplot(x='drought',y='ndvi',data=events_ndvi);
```
